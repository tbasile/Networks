```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd
```

```python

df = pd.read_csv('fifa.csv')
df.head(60)
```

```python
np.unique(df["rank_date"])[-800::]
```

```python
df_ult = df[df["rank_date"]=="2022-08-25"]
df_ult = df_ult[["rank","country_full"]]
df_ult = df_ult.sort_values(by=['country_full'])
df_ult.head(20)
```

```python
df_pen = df[df["rank_date"]=='1992-12-31']
df_pen = df_pen[["rank","country_full"]]
df_pen = df_pen.sort_values(by=['country_full'])
df_pen.head(20)
```

```python

def comparar(pais1, pais2):
    rank_ult_1 = df_ult[df_ult["country_full"]==pais1]["rank"].iloc[0]
    rank_ult_2 = df_ult[df_ult["country_full"]==pais2]["rank"].iloc[0]

    rank_pen_1 = df_pen[df_pen["country_full"]==pais1]["rank"].iloc[0]
    rank_pen_2 = df_pen[df_pen["country_full"]==pais2]["rank"].iloc[0]

    if (rank_ult_1 - rank_ult_2)*(rank_pen_1-rank_pen_2) < 0:
        return(1)
    else:
        return(0)
```

```python
paises = np.unique(df_ult["country_full"])

adj = np.zeros((len(paises),len(paises)))

    
    


for i in range(len(paises)):
    for j in range((len(paises))):
        if j == 0:
            print(i)
        adj[i][j] = comparar(paises[i],paises[j])
        
```

```python

df = pd.read_csv('tennis.csv')
df.head(60)
```

```python
np.unique(df["Date"])[-100::]
```

```python
df_ult = df[df["Date"]=='2019-09-16']
df_ult = df_ult[["Ranking","Player"]]
df_ult = df_ult.sort_values(by=['Player'])
df_ult
```

```python
df_pen = df[df["Date"]=='2017-07-31']
df_pen = df_pen[["Ranking","Player"]]
df_pen = df_pen.sort_values(by=['Player'])
df_pen
```

```python
nombres = np.intersect1d(np.unique(df_pen["Player"]),np.unique(df_ult["Player"]))
len(nombres)
```

```python
def comparar(nom1, nom2):
    rank_ult_1 = df_ult[df_ult["Player"]==nom1]["Ranking"].iloc[0]
    rank_ult_2 = df_ult[df_ult["Player"]==nom2]["Ranking"].iloc[0]

    rank_pen_1 = df_pen[df_pen["Player"]==nom1]["Ranking"].iloc[0]
    rank_pen_2 = df_pen[df_pen["Player"]==nom2]["Ranking"].iloc[0]

    if (rank_ult_1 - rank_ult_2)*(rank_pen_1-rank_pen_2) < 0:
        return(1)
    else:
        return(0)
    
comparar("Albert Ramos-Vinolas","Alexander Zverev")
```

```python
adj = np.zeros((len(nombres),len(nombres)))

    
    


for i in range(len(nombres)):
    for j in range((len(nombres))):
        if j == 0:
            print(i)
        adj[i][j] = comparar(nombres[i],nombres[j])
        
```

```python
G =nx.from_numpy_matrix(adj)
nx.draw_circular(G)
```

```python
plt.plot(nx.degree_histogram(G),"-o")
```

```python
list(nx.selfloop_edges(G))
```

```python
pen = pen[: , 0]
ult = ult[: , 0]
print(pen)
print(ult)
```

```python
def comparar(i, j):
    rank_ult_i = ult[i]
    rank_ult_j = ult[j]
    
    rank_pen_i = pen[i]
    rank_pen_j = pen[j]

    if (rank_ult_i - rank_ult_j)*(rank_pen_i-rank_pen_j) < 0:
        return(1)
    else:
        return(0)
    
```

```python
adj = np.zeros((len(univ),len(univ)))

    
    


for i in range(len(univ)):
    for j in range((len(univ))):
        if j == 0:
            print(i)
        adj[i][j] = comparar(i,j)
        
```

```python
lon = len(pen)
lon
```

```python
def comparar(i, j):
    rank_ult_i = ult[i]
    rank_ult_j = ult[j]
    
    rank_pen_i = pen[i]
    rank_pen_j = pen[j]

    if (rank_ult_i - rank_ult_j)*(rank_pen_i-rank_pen_j) < 0:
        return(1)
    else:
        return(0)
    
adj = np.zeros((lon,lon))

    
    


for i in range(lon):
    for j in range(lon):
        if j == 0:
            print(i)
        adj[i][j] = comparar(i,j)
        
```
